```python

#How do I filter rows of a pandas DataFrame by column value
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
# Filter the DataFrame rows to only show movies with a 'duration' of at least 200 minutes.


boolean=[]
for i in movies.duration: #for object in movies.duration
    if i >=200 :
        boolean.append(True)
    else:
        boolean.append(False)
boolean=pd.Series(boolean)
print(boolean.head())
```

```python
# How do I apply MULTIPLE filter criteria to a pandas DataFrame?
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
movies[movies.duration>=200].genre.head()
```

```python
#for applying filter on multiple columns using boolean operation , AND, OR
movies[(movies.duration>=200) & (movies.genre=='Drama')].head() #() - because to tell pandas order in which to filter
```

```python
#inside of bracket
(movies.duration>=200) & (movies.genre=='Drama')
```

```python

#movies - crime, drama, action- >=200 min
movies[(movies.duration>=200) & (movies.genre=="Drama") |(movies.genre=='Crime')]
```

```python

movies[(movies.duration>=200) & (movies.genre=="Drama") |(movies.duration>=200) & (movies.genre=='Crime')]
```

```python
#WASn't that lengthy?
movies[movies.duration>=200].genre.isin(['Drama',"Crime"])
```
