```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,3*np.pi,0.1)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sinx and cosinex')
plt.legend(['sinx','cosinex'])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
```
